```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("data-export (1).csv")
```

```python
df.head()
```

```python
# Set first row as column names
df.columns = df.iloc[0]

# Remove the header row from data and reset index
df = df.drop(index=0).reset_index(drop=True)

# Rename columns explicitly
df.columns = [
    "channel group",
    "DateHour",
    "Users",
    "Sessions",
    "Engaged sessions",
    "Average engagement time per session",
    "Engaged sessions per user",
    "Events per session",
    "Engagement rate",
    "Event count"
]
```

```python
df.info()
```

```python
df.head()
```

```python
# Convert DateHour to datetime
df["DateHour"] = pd.to_datetime(
    df["DateHour"],
    format="%Y%m%d%H",
    errors="coerce"
)

# Select numeric columns (exclude non-numeric ones)
numeric_cols = df.columns.drop(["channel group", "DateHour"])

# Convert selected columns to numeric
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")

# Extract hour from DateHour
df["Hour"] = df["DateHour"].dt.hour
```

#Engaged vs non Engaged sessions

```python
df.head()
```

```python
# Group sessions by channel
session_df = (
    df.groupby("channel group")[["Sessions", "Engaged sessions"]]
    .sum()
    .reset_index()
)

# Calculate non-engaged sessions
session_df["Non-Engaged"] = (
    session_df["Sessions"] - session_df["Engaged sessions"]
)

# Melt dataframe for plotting
session_df_melted = session_df.melt(
    id_vars="channel group",
    value_vars=["Engaged sessions", "Non-Engaged"]
)

# Plot
plt.figure(figsize=(8, 5))

sns.barplot(
    data=session_df_melted,
    x="channel group",
    y="value",
    hue="variable"
)

plt.title("Engaged vs Non-Engaged Sessions")
plt.xticks(rotation=45)

plt.show()
```

#Traffic by hour and channel

```python
df.head()
```

```python
# Prepare heatmap data
heatmap_data = (
    df.groupby(["Hour", "channel group"])["Sessions"]
    .sum()
    .unstack()
    .fillna(0)
)

# Plot heatmap
plt.figure(figsize=(12, 6))

sns.heatmap(
    heatmap_data,
    cmap="YlGnBu",
    linewidths=0.5,
    annot=True,
    fmt=".0f"
)

plt.title("Traffic by Hour and Channel")
plt.xlabel("Channel Group")
plt.ylabel("Hour of Day")

plt.show()
```

Is there any correlation between high traffic (sessions) and high engagement rate over time?

```python
# Prepare data for plotting
df_plot = (
    df.groupby("DateHour")[["Engagement rate", "Sessions"]]
    .mean()
    .reset_index()
)

# Plot
plt.figure(figsize=(10, 5))

plt.plot(
    df_plot["DateHour"],
    df_plot["Engagement rate"],
    label="Engagement Rate"
)

plt.plot(
    df_plot["DateHour"],
    df_plot["Sessions"],
    label="Sessions"
)

plt.title("Engagement Rate vs Sessions Over Time")
plt.xlabel("DateHour")
plt.legend()
plt.grid(True)

plt.show()
```
